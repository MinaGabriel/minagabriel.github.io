
# Quick Sort Introduction

Quick Sort is a divide-and-conquer algorithm that sorts items by partitioning the array into subarrays around a pivot element. The steps are as follows:

1. **Choose a pivot element** from the array.
2. **Partition** the array into two subarrays:
    - Elements less than or equal to the pivot.
    - Elements greater than the pivot.
3. **Recursively apply** the above steps to the subarrays.
4. **Combine** the sorted subarrays and the pivot into a sorted array.

The efficiency of Quick Sort depends on the partition strategy used. There are several partition methods, but we'll focus on three: Lomuto & Hoare.
